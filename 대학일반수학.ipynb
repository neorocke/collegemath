# Chapter 5,6 극대 극소 도함수 이계도함수

```python
# from sympy import symbols
from sympy import *
import numpy as np
```

```python
# 평균 변화율
def average(a,b):
    m = max(a,b)
    n = min(a,b)

    x = symbols('x')

    fx = x**2 + 2 * x +4
    fb = fx.subs(x, m)
    fa = fx.subs(x, n)

    result = (fb - fa) / (b -a)
    return result

print(average(0,2))
```

```python
# 극대값 극소값
def maxmin(fprime):
    ans_list = []
    for i in range(len(fprime)):
        ans_tuple = ()
        ans_tuple = ans_tuple + (fprime[i], fx.subs(x, fprime[i]))
        ans_list.append(ans_tuple)
    return ans_list
```
